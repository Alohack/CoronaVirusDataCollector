### 2 
Create a (40,10) random numpy array. You can create a (0-100) number and take the floor() of it. Now add a vector of 0-s from above, below, left and right - like a frame for your array. The resulting array is a (42,12) array. Print it.

```python
#I will need this function later in the 12th task

def add_zeros(arr):
    number_of_rows = arr.shape[0]
    number_of_cols = arr.shape[1]
    
    # inserting 0s from above below left and right
    arr = np.insert(arr, (0,number_of_rows), 0, axis=0)
    arr = np.insert(arr, (0,number_of_cols), 0, axis=1)
    
    return arr
```

```python

number_of_rows = 40
number_of_cols = 10

# creating random array
random_array = np.random.rand(number_of_rows, number_of_cols)
# making sure that each number is in [0;100] and taking the floor of each number
array_from_2 = np.floor(random_array*100).astype(int)

# calling the function wich will add zros
array_from_2 = add_zeros(array_from_2)

print(array_from_2)
```

### 3 
Take the array from 2, and horizontally split it into 2 equal parts, then add a column of np.ones() in between, and stack the resulting 3 objects back together. Print the resulting.

```python
number_of_rows = array_from_2.shape[0]
number_of_cols = array_from_2.shape[1]

# splitting the array horizontally
horizontal_split = np.hsplit(array_from_2, 2)

# creating column of ones and inserting it to the horizontal split
column_of_ones = np.ones((number_of_rows,1)).astype(int)
horizontal_split.insert(1, column_of_ones)

# stacking the resulting split
array_from_3 = np.concatenate(horizontal_split, axis = 1)

print(array_from_3)
```

### 4

```python
number_of_rows = array_from_3.shape[0]
number_of_cols = array_from_3.shape[1]

# splitting the array vertically into 3 parts
vertical_split = np.vsplit(array_from_3, 3)

# creating a row of twos and inserting it between parts
row_of_twos = np.full((1,number_of_cols),2)
vertical_split.insert(2,row_of_twos)
vertical_split.insert(1,row_of_twos)

# stacking the resulting split
array_from_4 = np.concatenate(vertical_split)

print(array_from_4)
```

```python
# Some backstory:
# There was a hypothesis that every continuous function at [a,b]
# must be differentiable at least in one point at (a,b)
# But there came Weierstrass and said "nope"
# So here is an examle of such function, which is continuous
# but it's not differentiable at any point

#it's graph is a fractal :D

a = 0.5 # a is a value between 0 and 1
b = 5 # b is a positive odd integer
N = 100 # N is some large value, since the sum is infinite
# since we have a**n in every element of the sum
# there is no need to sum up after N = 100

def weierstrass(x):
    y = 0;
    for n in range(0,N):
        y += (a**n)*np.cos( (b**n)*np.pi*x ) 
    return y
def plot_fun(f,xmin,xmax):
    x = np.linspace(xmin, xmax, 1000)
    plt.plot(x, f(x))

xmin = -2
xmax = 2

plot_fun(weierstrass, xmin, xmax)
```

### 12.* 
Make a smoother image by applying a filter 
- hint: use a convolution

```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img=mpimg.imread('images/noisy.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
import imageio

im = imageio.imread('images/noisy.png')
im = np.asarray(im)

im.shape
```

```python
def smooth_color(color, img_filter):
    number_of_rows = color.shape[0]
    number_of_cols = color.shape[1]
    
    # adding 2 rows of zeros from bottom and top
    # and 2 columns of zeros from left and right
    color_with_zeros = add_zeros(add_zeros(color))
    
    # this will contain our filtered image
    color_filtered = np.zeros(color.shape)
    
    for i in range(number_of_rows):
        for j in range(number_of_cols):
            # macing a convolution here
            color_filtered[i][j] = (img_filter*color_with_zeros[i:i+5, j:j+5]).sum()
    return color_filtered
```

```python
# the given filter
img_filter = 1/273 * np.array([
[ 1,  4,  7,  4, 1],
[ 4, 16, 26, 16, 4],
[ 7, 26, 41, 26, 7],
[ 4, 16, 26, 16, 4],
[ 1,  4,  7,  4, 1]
])

# checking if I wrote all correct from the slack :D
# the sum must be 1
img_filter.sum()
```

```python
number_of_rows = im.shape[0]
number_of_cols = im.shape[1]

# filtering the image
filtered_img = smooth_color(im[:,:,0], img_filter)
```

```python
# plotting 2 images, the original one and the filtered one
f, (ax1, ax2) = plt.subplots(1,2)

ax1.imshow(img)
ax2.imshow(filtered_img, cmap = 'gray')
```

```python
# calling the filtering function again on already filtered image
twice_filtered_img = smooth_color(filtered_img, img_filter)

# plotting the original, filtered and twice filtered images
f, (ax1, ax2, ax3) = plt.subplots(1, 3)

ax1.imshow(img)
ax2.imshow(filtered_img, cmap = 'gray')
ax3.imshow(twice_filtered_img, cmap = 'gray')
```

```python
# filtering the already filtered image 20 times more (This may take a while)
twenty_times_filtered = filtered_img
for i in range(10):
    twenty_times_filtered = smooth_color(twenty_times_filtered, img_filter)

# plotting the original, filtered and 20 times filtered images 
f, (ax1, ax2, ax3) = plt.subplots(1, 3)

ax1.imshow(img)
ax2.imshow(filtered_img, cmap = 'gray')
ax3.imshow(twenty_times_filtered, cmap = 'gray')
```
